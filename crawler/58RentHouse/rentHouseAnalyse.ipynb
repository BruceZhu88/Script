```python
from pyecharts import Bar
import sqlite3
from numpy import unique


def SQL(db_path, sql):
    conn = sqlite3.connect(db_path)
    conn.text_factory = str
    value = []
    c = conn.cursor()
    cursor = c.execute(sql)
    if ("select" or "SELECT") not in sql:
        conn.commit()
        conn.close()
    else:
        for row in cursor:
            value.append(row)
    conn.close()
    return value


def avg(region, room_type):
    prices = SQL(
        path, "select price from HOUSE where region='{}' and room_type='{}'".format(region, room_type))
    if len(prices) == 0:
        return 0
    avg = round(sum([int(d[0]) for d in prices]) / len(prices), 0)
    return avg


def sort_avg(region, room_type):
    avgs = [(r, avg(r, room_type)) for r in region if avg(r, room_type) > 0]
    avg_sorted = sorted(avgs, key=lambda item: item[1])
    region = [a[0] for a in avg_sorted]
    price = [a[1] for a in avg_sorted]
    return region, price

path = '58house.db'

regions = [d[0] for d in SQL(path, "select region from HOUSE") if d[0] != '']
uni_regions = unique(regions)
region_1_0_1, price_1_0_1 = sort_avg(uni_regions, '1室0厅1卫')
region_1_1_1, price_1_1_1 = sort_avg(uni_regions, '1室1厅1卫')
region_2_1_1, price_2_1_1 = sort_avg(uni_regions, '2室1厅1卫')
region_3_1_1, price_3_1_1 = sort_avg(uni_regions, '3室1厅1卫')
#for a in avg_sorted:
#    print('{} 平均租房价格:¥{}'.format(a[0], a[1]))

#均价
bar = Bar("Bar chart", "深圳所有地区均价")
bar.add("1室0厅1卫", region_1_0_1 ,price_1_0_1, mark_line=["average"], mark_point=["max", "min"])
import os
bar.render('{}/均价1室0厅1卫.html'.format(os.getcwd()))

bar = Bar("Bar chart", "深圳所有地区均价")
bar.add("1室1厅1卫", region_1_1_1 ,price_1_1_1, mark_line=["average"], mark_point=["max", "min"])
import os
bar.render('{}/均价1室1厅1卫.html'.format(os.getcwd()))

bar = Bar("Bar chart", "深圳所有地区均价")
bar.add("2室1厅1卫", region_2_1_1 ,price_2_1_1, mark_line=["average"], mark_point=["max", "min"])
import os
bar.render('{}/均价2室1厅1卫.html'.format(os.getcwd()))

bar = Bar("Bar chart", "深圳所有地区均价")
bar.add("3室1厅1卫", region_3_1_1 ,price_3_1_1, mark_line=["average"], mark_point=["max", "min"])
import os
bar.render('{}/均价3室1厅1卫.html'.format(os.getcwd()))


```
